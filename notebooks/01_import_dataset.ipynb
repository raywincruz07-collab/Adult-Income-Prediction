```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
census_income = fetch_ucirepo(id=20) 
  
# data (as pandas dataframes) 
X = census_income.data.features 
y = census_income.data.targets 
  
# metadata 
print(census_income.metadata) 
  
# variable information 
print(census_income.variables) 
```

```python
import pandas as pd
import os

# Create the data folder if it doesn't exist
os.makedirs("data", exist_ok=True)

# Combine features and target into one dataframe
census_income_df = pd.concat([X, y], axis=1)

# Save dataset in the data folder
census_income_df.to_csv(os.path.join("data", "census_income_full.csv"), index=False)

print("Saved census_income_full.csv in the data directory")
```
